```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModel
```

```python
ratings = pd.read_csv(RATINGS_PATH)
meta_content = pd.read_csv(META_PATH,usecols=['id','title','overview'])#1997-08-20
to_drop = ["1997-08-20","2012-09-29","2014-01-01"]
for drop_error in to_drop:
    meta_content = meta_content[meta_content.id != drop_error]
meta_content.id = meta_content.id.astype('int64')
movies_ids = list(set(ratings.movieId.tolist()))
```

```python
mask = meta_content['id'].isin(movies_ids)
meta_content_filtered = meta_content.loc[mask]
meta_content_filtered=meta_content_filtered.set_index('id').dropna()
```

```python
def get_overview_embeddings(meta_content):
    tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
    model = AutoModel.from_pretrained("bert-base-uncased")
    cols = {"id","title","overview","vector"}
    df = pd.DataFrame(columns=cols)
    for index, row in tqdm(meta_content.iterrows()):
        title = row['title']
        overview = row['overview']
        text = title+" : "+ overview
        inputs = tokenizer(text,return_tensors="pt",padding=True, truncation=True)
        outputs = model(**inputs)
        emb = outputs[1].squeeze(dim=0).shape
        tmp = pd.DataFrame({"id":index,"overview":overview,"title":title,"vector":emb})
        df=df.append(tmp)
    #inputs = tokenizer(meta_content, return_tensors="pt")
    return df
    #
```

```python
df = get_overview_embeddings(meta_content_filtered)
```
